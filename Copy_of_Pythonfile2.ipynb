```python
price_of_burger = 150
price_of_pizza = 450
price_of_coke = 80
price_of_latte = 250
price_of_pastry = 140
Total_price = price_of_burger + price_of_coke + price_of_pizza + price_of_latte +price_of_pastry
discount = Total_price * 10/100
payable_bill = Total_price-discount
print("price_of_burger = ", price_of_burger)
print("price_of_pizza  = ", price_of_pizza)
print("price_of_coke   = ", price_of_coke)
print("price_of_latte  = ", price_of_latte)
print("price_of_pastry = ", price_of_pastry)
print("------------------------")
print("Total bill      = ", Total_price)
print("Discount = ", discount)
print("payable bill = ", payable_bill)
```
